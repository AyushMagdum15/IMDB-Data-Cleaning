```python
#creating a function to covert the string values to list with help of 'eval' function
def convert_to_lst(string):
    try:
        return eval(string)
    except:
        return np.nan
```

```python

df['genres']=df['genres'].apply(convert_to_lst)
```

```python
type(df['genres'].iloc[0])
```

```python
df['crew']=df['crew'].apply(convert_to_lst)
```

```python
type(df['crew'].iloc[0])
```

```python
df['cast']=df['cast'].apply(convert_to_lst)
```

```python
#making a copy of original DataFrame
df1=df.copy()
```

```python
#creating a new column as profit 
df1['Profit']=df['revenue']-df['budget']
```

```python
#calculating the max profit
max_profit=df1['Profit'].max()
```

```python
max_profit_film=df1[df1['Profit']==max_profit].reset_index(drop=True)
```
